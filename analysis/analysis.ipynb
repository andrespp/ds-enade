### a2 Seaborn distplot

```python
df.columns
```

```python
dfa2 = df[['NU_ANO', 'NM_GRUPO','NM_INSCRICAO','NM_PRES','NT_GER','NT_FG','NT_CE']]
```

```python
# Médias
for year in dfa2['NU_ANO'].unique():
    print('{}: {:4.6f}  {:4.6f}  {:4.6f}'.format(
        year, 
        dfa2[dfa2['NU_ANO']==year]['NT_GER'].mean(),
        dfa2[dfa2['NU_ANO']==year]['NT_FG'].mean(),
        dfa2[dfa2['NU_ANO']==year]['NT_CE'].mean()))
```

```python
#f, axes = plt.subplots(1, 2)
len(dfa2['NU_ANO'].unique())
sns.distplot(dfa2['NT_GER'])

dfa2['NT_GER'].mean()
```

**Médias**

```python
# Graph is 3xN (compute N considering the number of distinct years in the dataset)
ncols = 3
nrows = int(len(dfa2['NU_ANO'].unique())/3)
if (len(dfa2['NU_ANO'].unique())%3) != 0:
    nrows +=1

# Plot
f, axes = plt.subplots(nrows, ncols)
i = 0
j = 0
for year in dfa2['NU_ANO'].unique():
    sns.distplot(dfa2[dfa2['NU_ANO']==year]['NT_GER'], ax=axes[i,j])
    if j==2:
        i += 1
    j += 1
    if i >= nrows:
        break
    if j >= ncols:
        j = 0
```

### a3 Média - Série temporal

#### Ingressantes e concluintes

```python
# Todos
dfa3 = df[['NU_ANO', 'NT_GER','NT_FG','NT_CE']].groupby(['NU_ANO']).mean()
```

```python
dfa3 = dfa3.iloc[1:]

dfa3.index
```

```python
# Create Figure (empty canvas)
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)

# Plot on that set of axes
axes.plot(dfa3.index, dfa3['NT_GER'], 'r-o', label='Nota Global')
axes.plot(dfa3.index, dfa3['NT_FG'], 'g-s', label='Formação Geral')
axes.plot(dfa3.index, dfa3['NT_CE'], 'b-^', label='Componente Específico')
axes.set_xlabel('Ano da Avaliação')
axes.set_ylabel('Nota')
axes.set_title('Médias ENADE por ano - Ingressantes e Concluintes')
axes.legend(loc=3)

plt.show()
```

#### Ingressantes

```python
# Ingressantes
dfa3in = df[df['NM_INSCRICAO']=='Ingressante']
dfa3in = dfa3in[['NU_ANO', 'NT_GER','NT_FG','NT_CE']].groupby(['NU_ANO']).mean()
dfa3in
```

```python
# Create Figure (empty canvas)
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)

# Plot on that set of axes
axes.plot(dfa3in.index, dfa3in['NT_GER'], 'ro', label='Nota Global')
axes.plot(dfa3in.index, dfa3in['NT_FG'], 'go', label='Formação Geral')
axes.plot(dfa3in.index, dfa3in['NT_CE'], 'bo', label='Componente Específico')
axes.set_xlabel('Ano da Avaliação')
axes.set_ylabel('Nota')
axes.set_title('Médias ENADE por ano - Ingressantes')
axes.legend(loc=4)

plt.show()
```

#### Concluintes

```python
# Concluintes
dfa3out = df[df['NM_INSCRICAO']=='Concluinte']
dfa3out = dfa3out[['NU_ANO', 'NT_GER','NT_FG','NT_CE']].groupby(['NU_ANO']).mean()
```

```python
# Create Figure (empty canvas)
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)

# Plot on that set of axes
axes.plot(dfa3out.index, dfa3out['NT_GER'], 'r-o', label='Nota Global')
axes.plot(dfa3out.index, dfa3out['NT_FG'], 'g-s', label='Formação Geral')
axes.plot(dfa3out.index, dfa3out['NT_CE'], 'b-^', label='Componente Específico')
axes.set_xlabel('Ano da Avaliação')
axes.set_ylabel('Nota')
axes.set_title('Médias ENADE por ano - Concluintes')
axes.legend(loc=3)

plt.show()
```
